```python
import csv
from osgeo import gdal
import numpy as np
```

```python
def read_raster(file_path):  # 返回栅格值的数组
    dataset = gdal.Open(file_path)
    if dataset is None:
        print("无法打开文件: ", file_path)
        return None

    band = dataset.GetRasterBand(1)
    raster_array = band.ReadAsArray()

    return raster_array

def read_multiple_rasters(file_paths):   # 返回各栅格值数组的列表
    num_files = len(file_paths)
    raster_arrays = []

    for i in range(num_files):
        raster_array = read_raster(file_paths[i])
        if raster_array is not None:
            raster_arrays.append(raster_array)
    # print(raster_arrays)
    return raster_arrays

def save_to_csv(file_paths, output_file):
    # print('kaishi')
    raster_arrays = read_multiple_rasters(file_paths)
    num_files = len(raster_arrays)
    num_rows, num_cols = raster_arrays[0].shape  # 获取行数、列数
    # print(num_rows, num_cols)

    with open(output_file, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        # 写入表头，即各指数的名称
        writer.writerow(["Point"] + [index_list[i] for i in range(num_files)])
        # 写入每个栅格点的值
        for i in range(num_rows):
            for j in range(num_cols):
                values_at_point = [raster_array[i][j] for raster_array in raster_arrays]
                writer.writerow([f"({i}, {j})"] + values_at_point)
                
def check_xy(file_paths):   # 检查行列数
    raster_arrays = read_multiple_rasters(file_paths)
    num_files = len(raster_arrays)
    for i in range(0,8):
        num_rows, num_cols = raster_arrays[i].shape   # 获取行数、列数
        # print(num_rows, num_cols)
```
